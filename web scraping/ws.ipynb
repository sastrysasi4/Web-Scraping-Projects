```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

import urllib3
urllib3.disable_warnings()

data=[]
html_text= requests.get("https://hmfw.ap.gov.in/covid_dashboard.aspx",verify=False).text
soup = BeautifulSoup(html_text,"lxml")

table = soup.find('table', attrs={'class':'table table-bordered table-striped'}) 
rows = table.find_all('tr')

for row in rows:
    cols = row.findAll('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) 
    
df= pd.DataFrame(data,columns=['District','ConfirmedCases','Cured','Deceased'])
df=df.drop(df.index[[0]])
df
```
